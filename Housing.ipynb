Data visualization

```python
sns.lineplot()
```

```python
sns.barplot( y='price' , hue='floors',  data=df )
plt.title('Price by Floors')
plt.grid(True)
plt.xlabel('Floor')
plt.xticks()
plt.show()
```

```python
# Compare price distributions for different floor counts
import matplotlib.ticker as ticker
sns.kdeplot(data=df, x='price', hue='floors', fill=True)
plt.title('Price Distribution by Number of Floors')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks()
plt.show()
```

```python
sns.barplot( y='price' , hue='bathrooms',  data=df )
plt.title('Price by the number of bathrooms')
plt.grid(True)
plt.xlabel('Bathrooms')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks()
plt.show()
```

From the plot above we can conclude that the more bedrooms a house has the more expensive it is.

```python
sns.barplot( y='price' , hue='bedrooms',  data=df )
plt.title('Price by the number of bedrooms')
plt.grid(True)
plt.xlabel('Bedrooms')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks()
plt.show()
```

```python
import matplotlib.ticker as ticker
sns.barplot( y='price' , hue='waterfront',  data=df ,palette='muted' )
plt.title('Price by the number of bedrooms')
plt.grid(True)
plt.xlabel('Waterfront')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks()
plt.show()
```

The plot above illustrates that houses with waterfront cost more than those without a waterfront

```python

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, y='price', x='sqft_living', fill=True, color='blue')
plt.title('House Price Distribution')
plt.show()
```
